```python
def rng(seed: int, n: int) -> List[float]:
    nums = [seed]
    m = 104_729
    a = 70_000
    b = 15_000
    for _ in range(n):
        nums.append(
            (nums[-1] * a + b) % m
            )
    return list(map(lambda x: x / m, nums[1:]))
```

```python
import time
```

```python
nums = rng(int(time.time()), 10)
nums
```

# Atributos

```python
class Test:
    def __init__(self, attr):
        self.attr = attr

    @property
    def attr2(self):
        return self.attr * 2
```

```python
x = Test(attr=3)
```

```python
x.attr
```

```python
x.attr2
```
