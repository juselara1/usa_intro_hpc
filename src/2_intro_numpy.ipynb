```python
class Complex:
    r: float
    im: float

    def __init__(self, r: float, im: float):
        self.r = r
        self.im = im

    def __repr__(self) -> str:
        return f"{self.r} + {self.im}·i"

    def __str__(self) -> str:
        return self.__repr__()

    def __add__(self, b: "Complex") -> "Complex":
        return Complex(r=self.r + b.r, im=self.im + b.im)
```
